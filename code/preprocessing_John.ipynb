```python
# Calculating the average number of films produced in a year given that they produced at least one film

us_data['year'] = pd.DatetimeIndex(us_data['release_date']).year
in_date = us_data[(us_data['year'] >= 2000) & (us_data['year'] <= 2010)]
df = pd.DataFrame()

#count the number of films each company produced in each year and add to dataframe
for year in range(2000, 2008):
    counter = in_date[in_date['year'] == year]['first_company'].value_counts()
    
    for i in range(len(counter)):
        df.loc[year, counter.index[i]] = counter[i]

# drop if only made one film the entire time because that might be a shell company
no_shell_companies = df.loc[:, (df.sum(axis=0) > 1)]

#calulate mean for each row
df['mean'] = df.mean(axis=1)
no_shell_companies['mean'] = no_shell_companies.mean(axis=1)

print('average with potential for shell companies: {}'.format(np.mean(df['mean'])))
print('average without potential shell companies: {}'.format(np.mean(no_shell_companies['mean'])))
```

```python
def gender_score(cast_list, n):
    gender_scores = np.zeros(len(cast_list))
    
    #loop through each cast list
    for i in range(len(cast_list)):
        count_female = 0

        #loop through actors in cast list
        for j in range(n):
            try:
                if cast_list[i][j]['gender'] == 1:
                    count_female = count_female + 1
            except:
                pass
        
        gender_scores[i] = count_female / n
            
    return gender_scores

us_data['gender_score'] = gender_score(us_data['cast'], 5)
```

```python
oscars = pd.read_csv('data/oscars.csv')
```

```python
def has_oscar_winner(oscar_list, df):
    cast_list = df['cast']
    has_oscar_winner = np.zeros(len(cast_list))
    winner_list = oscar_list['name'].to_list()
    normalized_list = [name.lower().strip() for name in winner_list]
    
    #loop through cas list
    for i in range(len(cast_list)):
        has_winnner = False
        
        #loop through actors in cast list
        for j in range(len(cast_list[i])):
#             print(len(cast_list), len(cast_list[i]))
            if cast_list[i][j]['name'].lower().strip() in normalized_list:
                has_winnner = True
            
        has_oscar_winner[i] = has_winnner
                
    return has_oscar_winner
```

```python
train_sub_1980_2012['has_oscar_winner'] = has_oscar_winner(pre_2013_winners, train_sub_1980_2012)
val_2013_2014['has_oscar_winner'] = has_oscar_winner(pre_2013_winners, val_2013_2014)
train_1980_2014['has_oscar_winner'] = has_oscar_winner(pre_2015_winners, train_1980_2014)
test_2015_2017['has_oscar_winner'] = has_oscar_winner(pre_2015_winners, test_2015_2017)
```

```python
train_sub_1980_2012.to_csv('data/Current/train_sub_1980_2012_prod_comp_class_actor_director_istopk_logrev_genres.csv')
train_1980_2014.to_csv('data/Current/train_1980_2014_prod_comp_class_actor_director_istopk_logrev_genres.csv')
test_2015_2017.to_csv('data/Current/test_2015_2017_prod_comp_class_actor_director_istopk_logrev_genres.csv')
val_2013_2014.to_csv('data/Current/val_2013_2014_prod_comp_class_actor_director_istopk_logrev_genres.csv')
```

```python
def num_oscar_winners(cast_list):
    num_winner_holder = np.zeros(len(cast_list))
    winner_list = oscars['name'].to_list()
    normalized_list = [name.lower().strip() for name in winner_list]
    
    #loop through all cast lists
    for i in range(len(cast_list)):
        num_winners = 0
        
        #loop through all actors in cast list
        for j in range(len(cast_list[i])):
            #if also in list of oscar winners
            if cast_list[i][j]['name'].lower().strip() in normalized_list: 
                num_winners =+ 1
            
        num_winner_holder[i] = num_winners
                
    return num_winner_holder
```

```python
us_data['num_oscar_winners'] = num_oscar_winners(us_data['cast'])
```
