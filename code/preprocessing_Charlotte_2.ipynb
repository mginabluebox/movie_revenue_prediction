```
movie = pd.read_csv(path+'aggregated_1119.csv')
movie.id = movie['id.1.1'].astype('int')
movie.drop(['id.1','id.1.1'],axis = 1,inplace=True)
movie.head()
```

```
# convert list 
prd_countries = movie['production_countries'].tolist()
â€©for i in movie[movie['production_countries'].notnull()].index: 
    if type(movie['production_countries'][i]) is not list:
        prd_countries[i] = ast.literal_eval(movie['production_countries'][i])
```

```
prd_count = pd.DataFrame(prd_countries)
prd_count['id'] = movie['id']
```

```
usa = []
for idx in prd_count.index:
    mov = prd_count.loc[idx]
    count = 0
    for i in mov.index:
        if type(mov[i]) == dict:
            if mov[i]['iso_3166_1'] == 'US':
                count +=1
    if count > 0:
        usa.append(True)
    else: usa.append(False)
movie['usa'] = usa
movie_us_only = movie.iloc[usa]
```

```
temp = original.set_index('id').loc[movie_us_only.set_index('id').index][['adult','video','revenue']]
movie_us_only.reset_index(inplace=True,drop=True)
```

```
movie_us_only['log_revenue_num'] = movie_us_only['revenue']
movie_us_only['adult'] = temp['adult'].values
movie_us_only['video'] = temp['video'].values
movie_us_only['revenue'] = temp['revenue'].values
```

```
movie_us_only['adult1'] = temp['adult']
movie_us_only['video1'] = temp['video'].values
```
